```python
import matplotlib.pyplot as plt
import numpy as np
import prettyPlots
prettyPlots.setStyle()
```

```python
x = np.linspace(-2,2,100)
y = np.exp(x)
```

```python
plt.plot(x,y)
plt.plot(0,np.exp(0),'o')
plt.plot(1,np.exp(1),'o')
plt.text(1.1,np.exp(1)*0.95,'$f(1) = e$',size=20)
plt.text(-1,np.exp(0)*1.1,'$f(0) = 1$',size=20)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.savefig('/Users/paha3326/miniconda3/envs/planets/planetary-atmospheres/exponential.pdf',bbox_inches='tight')
```
