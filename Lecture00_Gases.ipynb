## Basic Properties of Gases

### Particle nature of gases

Gases are made up of many particles, which are usually molecules (e.g., N$_2$, O$_2$, CO$_2$), but they can also be monatomic (e.g., noble gases like Ar and Xe). Gases quickly fill their containers, due to random motions in all directions. Each molecule in a gas posseses its own kinetic energy, but together they are an *ensemble* with a well-defined average kinetic energy. Temperature is a measure of this energy, and all gases within the same container have the same temperature in equilibrium. For each *degree of freedom* (read: dimension, or direction of motion), the average kinetic energy of the ensemble is $E_i = k_\mathrm{B}T/2$. So, in three dimensions, the average kinetic energy is

\begin{align}
    E = E_x + E_y + E_z = \frac{3}{2}k_\mathrm{B}T
\end{align}

A more complete description of velocities in a gas is derived from kinetic theory and can be modeled by the [Maxwell-Boltzmann Distribution](https://en.wikipedia.org/wiki/Maxwell%E2%80%93Boltzmann_distribution). The average of the M-B distribution is the thermal velocity, which we find by setting the kinetic energy of a particle equal to the average kinetic energy of the ensemble:

\begin{align}
    \frac{1}{2}mv_\mathrm{th}^2 &= E \\
    \\
    \rightarrow v_\mathrm{th}^2 &= \sqrt{\frac{2E}{m}} = \sqrt{\frac{3k_\mathrm{B}T}{m}}
\end{align}

Each particle also has a certain size, with radius $r \sim 2~\mathring{\mathrm{A}} = 2\times 10^{-10}~\mathrm{m}$. Particles in the gas are in constant motion, and therefore collide with one another. The cross section for mutual collisions is given by $\sigma_\mathrm{c} \sim \pi d^2 = 4\pi r^2$, where $d = 2r$ is the particle diameter. (Note: the collision cross section is *not* just the cross-sectional area of one molecule, because the two particles touch when they are separated by a distance $2r$.) The mean free path is the typical distance traveled between collisions, and is given by $\ell = (n\sigma_\mathrm{c})^{-1}$.

### Number vs. number density vs. mass density

In a volume of space $V$ there are a certain number of gas molecules $N$. If there are several components (e.g., CO$_2$ and N$_2$), we call their numbers $N_i$, and the total number is $N = N_1 + N_2 + N_3 + ... = \sum_i N_i$. The *number density*, $n$ is defined as the number of particles per unit volume. In the same way as we did for the number of particles, we define number *density* for both the individual components ($n_i = N_i/V$) and the total amount of gas in the volume $V$:

\begin{align}
    n &= \frac{N}{V} = \frac{N_1+N_2+N_3+...}{V} = n_1 + n_2 + n_3 + ... \\
      &= \sum_i n_i
\end{align}

Each particle has its own mass. To find the mass density of the volume of gas with number density $n$, we simply multiply by the mass of an individual particle:

\begin{align}
    \mathrm{Mass~density} = \rho = mn
\end{align}

In the case where the gas is made up of several different types of particles with masses $m_i$, the mass densities of each component are given by

\begin{align}
    \rho_i = m_i n_i
\end{align}

and the total mass density is

\begin{align}
    \rho = \sum_i \rho_i = \sum_i m_i n_i = \frac{m_1N_1 + m_2N_2 + ...}{V}
\end{align}

How do we find the mass of a particle? First, find the atomic mass of its parts. For example, molecular nitrogen (the most common gas in Earth's atmosphere) has the formula N$_2$, meaning that it is composed of two N atoms. Nitrogen has an atomic mass of 14 amu (atomic mass units, sometimes called *Daltons*, Da), because it is made of 7 protons and 7 neutrons. Next, we convert atomic mass to kilograms using the conversion 1 amu = $1.66\times 10^{-27}$ kg. So, in summary,

\begin{align}
    m_i = \mathrm{Mass~in~kilograms} = (\mathrm{Molecular~mass~in~amu})\times(1.66\times 10^{-27}~\mathrm{kg/amu})
\end{align}

For any gas molecule, we can look up its [molecular mass](http://www.bmrb.wisc.edu/metabolomics/mol_mass.php) and apply the formula above to find its mass in kg.

### Column abundance (surface density)

It is often useful to know the total amount of a gas above a patch of surface area. This is expressed in units of kilograms per square meter, or number of particles per square meter. If the gas has uniform density (not typical), $\rho$, then the *column density* is just $s = \rho z$ and the *column abundance* is $C = n z$, where $z$ is the vertical thickness of the gas. In the more likely case that the density (and number density) are variable with height above he surface, we have to write this as an integral

\begin{align}
    s &= \int_0^\infty \rho(z)dz \\
      &= \int_0^\infty mn(z)dz = m\int_0^\infty n(z)dz = mC
\end{align}

## Physics of Gases

### Equation of state: Ideal Gas Law

The *equation of state* (EOS) simply describes the relationship between temperature, pressure, and density of a substance. In the case of gases, we often use the *ideal gas law* or *perfect gas* EOS:

\begin{align}
    p = nk_\mathrm{B}T
\end{align}

We can rewrite this in terms of the density by substituting the expression for number density given above:

\begin{align}
    p = \frac{\rho k_\mathrm{B}T}{m}
\end{align}
where $m$ is the mass of the gas molecule. It is helpful to become intuitively familiar with the EOS. What happens to a gas' pressure when its temperature goes up? What happens to the pressure and temperature of an expanding gas?

### Hydrostatic balance

Pressure is the force per unit area on the "walls of the container", $p = F/A$. Even if there is no container, a gas exerts a force in all directions, which must be balanced. The most common balancing force in planetary atmospheres is gravity, where the balance is given by hydrostatic equilibrium. Another useful way to think about pressure is that it quantifies the energy per unit volume: $p = F/A = E/V$.

Using hydrostatic balance, we can rewrite the column mass

\begin{align}
    \frac{\mathrm{Mass}}{\mathrm{Area}} = \frac{\mathrm{Force/Area}}{\mathrm{Acceleration}} &= \frac{p}{g} \\
    &= \frac{1}{g}\int_0^\infty dp = \frac{1}{g}\int_0^\infty \rho(z)g~dz \\
    &= \int_0^\infty \rho(z)dz = s
\end{align}

### Partial pressures

Dalton's law says that the total pressure of a gas is just the sum of the *partial pressures* of its constituents:

\begin{align}
    p = p_1 + p_2 + p_3 +... = \sum_i p_i
\end{align}

Using the ideal gas EOS, we can rewrite Dalton's law in terms of the densities of the constituent gases:

\begin{align}
    p = \sum_i p_i = \sum_i n_i k_\mathrm{B} T = \sum_i \frac{\rho_i k_\mathrm{B} T}{m_i}
\end{align}

### Abundances: volume vs. mass mixing ratio
Note that the partial pressure of each constituent is proportional to the number ($N_i$) of that kind of particle: $p_i = (N_i/V)k_\mathrm{B}T$. Since all of the constituents have the same temperature and occupy the same volume, this means that the fractional abundance of each component is given by its abundance $N_i$, divided by the total number of particles. We write this *volume mixing ratio* as

\begin{align}
    q_i &= \frac{N_i}{N} = \frac{N_i}{\sum_i N_i} = \frac{N_i}{N_1+N_2+...} \\
        &= \frac{n_i}{n}
\end{align}

Alternatively, we sometimes want to know the *mass mixing ratio*, which is the fraction of the total mass made up of one constituent. So, the mass mixing ratio is

\begin{align}
    x_i &= \frac{\rho_i}{\rho} = \frac{m_i N_i}{\sum_i m_i N_i} \\
        &= \frac{m_i n_i}{\sum_i m_i n_i}
\end{align}

### Heat capacity

If we add some amount of heat $dQ$ to a gas, it will warm up by an amount $dT$. The change in temperature for the given heat added depends on the heat capacity of the gas, $c_p$ and its mass, $M = \rho V$:

\begin{align}
    dT = \frac{dQ}{c_p~M} = \frac{dQ}{c_p~\rho V}
\end{align}

The units of heat are joules (J) and the units of heat capacity are joules per kilogram per degree ($\mathrm{J~kg^{-1}~K^{-1}}$). Substances with large heat capacities (e.g., water) can store or release lots of heat by changing temperature a little bit. Conversely, substances with small heat capacities (e.g., dry air) must change temperature a lot in order to absorb or release the same amount of heat.
