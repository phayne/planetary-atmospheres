Motions in planetary atmospheres are governed fundamentally by force balance. Different forces are dominant in different situations, so usually the first step in tackling a problem in atmospheric dynamics is to identify the most and least important of these forces for a given state of the system. Once we identify the appropriate dynamical regime, we can solve the equations of motion (either analytically or numerically, i.e. with a computer) to describe the resultant flow. Fluid dynamics can be a complex and intimidating subject. Fortunately, most phenomena in planetary atmospheres can be accurately described and understood using a simplified set of equations appropriate to a particular regime.

## Fundamentals of rotating fluids

All planets rotate, as do their atmospheres. Fluid motions inside the atmosphere can be strongly affected by planetary rotation. First, we introduce the equations of motion for a fluid in a rotating reference frame. After identifying each force term, we look at ways to determine their relative importance in order to solve problems.

### Equations of motion in a rotating system

Standing on the surface of the planet or floating in a balloon in the atmosphere, we study atmospheric motions from within a rotating reference frame. In this case, the equations of motion include both the direct forces and the *inertial* (or 'fictitious') forces due to our rotating perspective. Newton's second law, $m\vec{a} = \vec{F}$ can be written

\begin{align}
    m\vec{a} ~~~~~&= ~~~~~~~~~~~~~(\mathrm{Direct~forces}) ~~~~~~~~~~~~~~~~~~~~ + ~~~~~~~~(\mathrm{Inertial~forces}) \\
    &= ~~~~~~~~~~~~~~~~~~~\vec{F}_\mathrm{direct} ~~~~~~~~~~~~~~~~~~~~~~~~~~~~ + ~~~~~~~ \vec{F}_\mathrm{inertial} ~~~~ \\
    &= ~~\vec{F}_\mathrm{gravity} + \vec{F}_\mathrm{pressure} + \vec{F}_\mathrm{friction}~~~~~~~~~~~+ ~~~ ~~~~\vec{F}_\mathrm{coriolis} + \vec{F}_\mathrm{centrifugal}
\end{align}
The composite first term ($\vec{F}_\mathrm{direct}$) on the right-hand side (RHS) contains all of the forces acting on the mass within the rotating system, including gravity, pressure, and friction. The second term ($\vec{F}_\mathrm{coriolis}$) on the RHS is the Coriolis force, which arises due to conservation of angular momentum, as described below. The last term ($\vec{F}_\mathrm{centrifugal}$) on the right-hand side is the centrifugal force, which acts in a direction perpendicular to the axis of rotation.

When we measure the downward force of "gravity" acting on a stationary object on the surface of the Earth, we are really measuring the sum of the gravitational force and the centrifugal force due to Earth's rotation. Therefore, we define $\vec{g}$ to include both forces: $m\vec{g} = \vec{F}_\mathrm{gravity} + \vec{F}_\mathrm{centrifugal}$.

### Coriolis effect

On a planet with constant rotation rate, we define an angular velocity vector, $\vec{\Omega}$ aligned with the planet's spin axis and with magnitude equal to its angular velocity: $\Omega = 2\pi/P$ (rad s$^{-1}$), with $P$ the planet's rotation period.

\begin{align}
    \vec{F}_\mathrm{coriolis} = -2m\vec{\Omega}\times\vec{v}
\end{align}
where $\vec{R} = (x,y,z)$ is the vector from the center of the planet to the mass $m$, and $\vec{v} = (u,v,w)$ is the velocity vector.
