Pressure generally decreases with altitude, due to the lower weight of the air column above. We can quantify this intuitive fact using the concept of **hydrostatic equilibrium**. Pressure profiles in planetary atmospheres can be quickly estimated using the hydrostatic equations, with some knowledge of temperature and composition.

Consider a thin layer with thickness $dz$, containing material (could be air, water, or rock) with density $\rho$ (Fig. 1). Drawing a circle to define a column with cross-sectional area $dA$, we have now encompassed a small amount of material with mass $dm = \rho~dA~dz$.

**Figure 1:** Hydrostatic equilibrium is the balance between the pressure difference across a layer, and its weight under gravity. The diagram depicts a single layer with density $\rho$. The layer's mass per unit area is $\rho~dz$ and its weight per unit area, $\rho g~dz$, is balanced by the pressure difference $dp$.

---

If we could place this small region on a scale, it would have a weight

\begin{equation}
    dF = dm~g = (\rho~dA~dz) g
\end{equation}

where $g$ is the gravitational acceleration. Pressure is a force per unit area, $dp = dF/dA$, so the pressure exerted by this layer is

\begin{equation}
    dp = \frac{(\rho~dA~dz) g}{dA} = -\rho g~dz
\end{equation}

where the minus sign indicates that pressure increases downward. This is the condition of **hydrostatic equilibrium**, where the pressure change across a layer balances its own weight under the planet's gravity.

### Incompressible fluids

For the simplest case, when $\rho = \mathrm{constant}$, the pressure at some depth $h$ can be found by adding up the contributions from each of the layers resting on top of it (Fig. 2).

```python
# Display Figure 2
Image(filename='/Users/paha3326/main/Teaching/ASTR3720/hayne/notes/hydrostatic_element2.png', width=400)
```

For such an **incompressible fluid**, the total pressure is the sum of all the layers, $p \approx \sum dp = -\rho g \sum dz$, such that

\begin{equation}
    p(h) = \rho g h
\end{equation}

where $h = \big| \sum dz \big|$ is the total depth of the fluid.

In geophysics, the pressure underneath a static load is called the **overburden** (Fig. 3). More specifically, it is the **lithostatic stress** when applied to rocks, or **hydrostatic stress** when applied to fluids (e.g., water or air). Deep in the Earth's interior, pressures are measured in giga-Pascals (1 GPa = 10$^9$ Pa), but atmospheric pressures are generally measured in units of "bars", where 1 bar = 10$^5$ Pa. You will also see pressures reported in atm (= 1.01325 bar), so be careful. A common laboratory pressure scale is the torr (= 1/760$^\mathrm{th}$ bar), which is equivalent to one "millimeter of mercury" (Why would pressure have units of length?).

### Compressible fluids

Using the convention that $z$ is positive upward, the pressure decreases for a positive increment $dz$ in altitude: $dp = -\rho g~dz$. Rearranging, we then have a differential equation for the variation of pressure with height:
\begin{equation}
    \frac{dp}{dz} = -\rho g
\end{equation}

This is called the **hydrostatic equation**. To get the pressure at a given height, $z$, we integrate

\begin{align}
    p(z) &= \int_0^p dp' \\
         &= p(0) + \int_0^{z'} \frac{dp'}{dz'}dz' \\
         &\approx p(0) - g\int_0^z \rho(z')dz'
\end{align}

where the approximation on the last line can be made when the height of the atmosphere is small compared to the radius of the planet, since $g \approx \mathrm{constant}$ (Why?).

The integral form of the hydrostatic equation is useful if we know the density profile, $\rho(z)$. Otherwise, we can solve the differential equation directly, if we have knowledge of the **equation of state** (EOS): $\rho = \rho(p,V,T)$, where $V$ and $T$ are volume and temperature, respectively. Gases are usually described by the familiar EOS (i.e., the **ideal gas law**),

\begin{equation}
    p = n k_\mathrm{B} T
\end{equation}

where $n$ is the number density (# molecules/m$^{3}$), and $k_\mathrm{B}$ is the Boltzmann constant. Given the mass of individual particles, $m$ (kg/molecule), the number density is $n = \rho/m$, and we have

\begin{equation}
    \rho = \frac{m}{k_\mathrm{B} T}p
\end{equation}

At a fixed temperature, we can then write the hydrostatic equation

\begin{equation}
    \frac{dp}{dz} = -\rho g = -\frac{m}{k_\mathrm{B} T}p g
\end{equation}

Rearranging and defining the **scale height** $H \equiv \frac{k_\mathrm{B} T}{mg}$,

\begin{equation}
    \frac{dp}{p} = -\frac{dz}{H}
\end{equation}

which we can integrate to find the pressure difference between two altitudes:

\begin{equation}
    \int_{p(z_1)}^{p(z_2)}\frac{dp}{p} = -\int_{z_1}^{z_2}\frac{dz}{H}
\end{equation}

For an **isothermal** region of the atmosphere (i.e., $T = \mathrm{constant}$), the integrals become

\begin{equation}
    \ln\left(\frac{p(z_2)}{p(z_1)}\right) = -\frac{z_2-z_1}{H}
\end{equation}

such that the pressure at the altitude $z_2$ is

\begin{equation}
    p(z_2) = p(z_1)e^{-(z_2-z_1)/H}
\end{equation}

An equivalent, and more commonly used formula gives the pressure at an altitude $z$, relative to the surface pressure, $p_0$:

\begin{equation}
    p(z) = p_0 e^{-z/H}
\end{equation}

A few examples below illustrate the concept of scale height.

```python
# Example: plot pressure as a function of altitude

# Import Python libraries
import numpy as np
from matplotlib import pyplot as plt
import prettyPlots as pp
pp.setStyle()

# Variables
z = np.arange(0,100,0.1) # altitude [km]
p_0 = 1.0 # surface pressure [1 bar = 10**5 Pa]
H = [5, 10, 15] # scale height [km]

# Calculate pressure profiles and plot results
for h in H:
    p = p_0*np.exp(-z/h)
    plt.plot(p, z, label='$H =$ {:d} km'.format(h))
    
plt.xlabel('Pressure (bar)')
plt.ylabel('Altitude (km)')
plt.axis([0,p_0,0,np.max(z)])
plt.legend(frameon=False)
plt.savefig('/Users/paha3326/main/Teaching/ASTR3720/hayne/notes/scale_height_examples.pdf',bbox_inches='tight')
plt.show()
```

### Column mass

The total pressure at the surface of a rocky planet is the force per unit area of the atmosphere "pressing" downward. Again assuming gravity $g \approx \mathrm{constant}$, the force is $F = M g$, with $M$ the total mass of the column. We call the mass per unit area, $\mu = M/A$, the **column mass** (kg m$^{-2}$), and the pressure is $p_0 = F/A = \mu g$. This gives a useful formula for the column mass,

\begin{equation}
    \mu = \frac{p_0}{g}
\end{equation}

If we have a measurement (or an estimate from other measurements) of the surface pressure $p_0$ on a planet with radius $R$, we can also estimate the mass of its atmosphere by $M_{\mathrm{atm}} \approx 4\pi R^2 p_0/g$. Alternatively, if we know the pressure at some altitude above the surface, and can determine the scale height based on temperature and composition, we can use the formulas above to derive surface pressure and total atmospheric mass.
