# Holding onto an Atmosphere

Gravity binds atmospheres to planets. A dominant competing effect is the thermal motion of gas molecules, which is determined by their temperature. In fact, temperature itself is a measure of the random motions of an ensemble of particles; higher temperature = higher velocities. If a significant fraction of gas molecules reach velocities higher than the *escape velocity*, the atmosphere will lose mass over time (or never form in the first place). This kind of atmospheric evaporation is called Jeans escape. We can therefore compare the strength of gravity to the thermal velocity to estimate the likelihood that a planet will retain an atmosphere.

## Gravitational Escape

A particle can escape the gravity well of a planetary body by achieving the *escape energy*, which is equal in magnitude to the gravitational potential energy, $\Phi$. For a particle of mass $m$ bound to a planet with mass $M$ and radius $R$, the escape energy is

\begin{equation}
    E_\mathrm{esc} = \left|\Phi(R)\right| = \frac{GMm}{R}
\end{equation}

### Escape Velocity

We can calculate the escape velocity, $v_\mathrm{esc}$ by setting the kinetic energy equal to the escape energy:

\begin{align}
    \mathrm{Kinetic~Energy} &= \mathrm{Escape~Energy} \\
    \\
    \frac{1}{2}mv_\mathrm{esc}^2 &= \frac{GMm}{R} \\
    \\
\end{align}
\begin{equation}
    \rightarrow v_\mathrm{esc} = \sqrt{\frac{2GM}{R}}
\end{equation}

(What is the escape velocity in terms of the *density* of the planet?)


### Thermal Velocity and Escape Parameter

How do we know what fraction of molecules reach escape velocity? First, we can look at their average *thermal velocity*, $v_\mathrm{th}$. For an ensemble of particles at temperature $T$, the mean kinetic energy of each particle due to random motions (in three dimensions) is given by

\begin{equation}
    \mathrm{\langle Kinetic~Energy \rangle} = \frac{3}{2}k_\mathrm{B}T
\end{equation}

where $k_\mathrm{B}$ is the Boltzmann constant ($\approx 1.38\times10^{-23}$m$^2$kg s$^{-2}$ K$^{-1}$). We set this equal to the kinetic energy of an individual particle to derive the mean thermal velocity,

\begin{align}
    \frac{1}{2}mv_\mathrm{th}^2 = \frac{3}{2}k_\mathrm{B}T \\
    \\
    \rightarrow v_\mathrm{th} = \sqrt{\frac{3k_\mathrm{B}T}{m}} \\
\end{align}

(Does this formula make sense?) As a basic check on atmospheric retention, we can make the comparison

\begin{equation}
    v_\mathrm{th} \stackrel{?}{>} v_\mathrm{esc} 
\end{equation}

If the mean thermal velocity is comparable to (or larger than) the escape velocity, then there is little hope of retaining an atmosphere! However, even if the mean thermal velocity is much lower than the escape velocity, there may be some gas particles that are moving much faster than the average. Therefore, we need to account for the full *velocity distribution*, which is given by the Maxwell-Boltzmann formula:

\begin{equation}
    \mathrm{Probability~Distribution~Function~(P.D.F.)} = f(v) = \left(\frac{m}{2\pi k_\mathrm{B} T}\right)^{3/2} 4\pi v^2 e^{\frac{-mv^2}{2k_\mathrm{B}T}}
\end{equation}

The probability of a gas particle having velocity between $v$ and $v+dv$ is given by $f(v)dv$. For atmospheric escape, we are interested in the higher velocity end of the distribution. As $v\rightarrow \infty$, the PDF is dominated by the exponential term:

\begin{equation}
    \mathrm{Probability~of~Velocity}~v = f(v)dv \sim e^{\frac{-mv^2}{2k_\mathrm{B}T}}dv
\end{equation}

In the exponential term, we recognize the numerator as the kinetic energy of a particle with velocity $v$. Setting $v = v_\mathrm{esc}$, the kinetic energy is equal to the escape energy ($E_\mathrm{esc}$). Therefore, the probability of a particle reaching escape velocity can be written

\begin{equation}
    \mathrm{Probability~of~escape} \sim e^{-\frac{E_\mathrm{esc}}{k_\mathrm{B}T}}dv = e^{-\lambda}dv
\end{equation}

where we define the *escape parameter*: 
\begin{equation}
    \lambda \equiv \frac{E_\mathrm{esc}}{k_\mathrm{B}T}
\end{equation}

The escape parameter is approximately the ratio of the escape energy (gravitational binding energy) to the mean kinetic energy of the gas. Note that a large value of the escape parameter implies $E_\mathrm{esc} \gg \frac{3}{2}k_\mathrm{B}T$, or in words: the energy needed to escape is much larger than the typical kinetic energy of the gas. Therefore (maybe counterintuitively!), large escape parameter $\rightarrow$ small probability of escape.
